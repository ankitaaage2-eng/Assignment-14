```python
import pandas as pd

# Load dataset
df = pd.read_csv('ford_car_dataset.csv')

# Separate X and Y
X = df.drop('price', axis=1)
y = df['price']               # Dependent - Target

# Print shape
print("Shape of X:", X.shape)
print("Shape of Y:", y.shape)
print("\nFirst 5 rows of X:")
print(X.head())
```

```python
# Find categorical columns
cat_cols = X.select_dtypes(include='object').columns
print("Categorical columns:", list(cat_cols))

# One-Hot Encoding
X = pd.get_dummies(X, columns=cat_cols, drop_first=True)

# Convert to integer
X = X.astype(int)

print("\nFirst 5 rows after encoding:")
print(X.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Apply StandardScaler
scaler = StandardScaler()

# Numerical columns to scale
num_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

X[num_cols] = scaler.fit_transform(X[num_cols])

print("First 5 rows after scaling:")
print(X.head())
```

```python
from sklearn.model_selection import train_test_split

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Print shapes
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

# Create and train model
model = LinearRegression()
model.fit(X_train, y_train)

# Print intercept and coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

```python
# Predict on test data
y_pred = model.predict(X_test)

# First 10 predictions
print("First 10 Predicted values:")
print(y_pred[:10])

# First 10 actual values
print("\nFirst 10 Actual values:")
print(y_test.values[:10])
```

```python
from sklearn.metrics import r2_score

# Calculate R2 Score
r2 = r2_score(y_test, y_pred)
print("R2 Score:", r2)

# Interpretation
print("\nInterpretation: The model explains", round(r2*100,2), "% of the variance in car price.")
print("Comment: The model is performing well as R2 score is above 0.8")
```

```python
import joblib

# Save model
joblib.dump(model, "LR_ford_car.pkl")

# Save scaler
joblib.dump(scaler, "scaler.pkl")

# Save columns
joblib.dump(list(X.columns), "columns.pkl")

print("All 3 files saved: LR_ford_car.pkl, scaler.pkl, columns.pkl")
```
